## ノック6 : 繰り返し処理で新しいデータを追加する

```python
import os

os.listdir('data')
```

```python
from glob import glob

diff_files = glob('data/diff_*.csv')

print( diff_files )
```

```python
diff_files.sort()

print( diff_files )
```

```python
diff = pd.read_csv(diff_files[0], encoding='shift-jis', header=None, dtype=object)

print( len(diff) )

print( diff.head() )
```

```python
diff.columns = columns

print( diff.head(3) )

diff = diff.loc[diff['prefectureName'] == '静岡県']

print( diff.head(3) )
```

```python
data_test = data

print( len(data_test) )
print( len(data_test) == len(data) )

print( len(diff) )

data_test = pd.concat([data_test, diff], axis=0 )

print( len(data_test) )

data_test.tail(3)
```

```python
for f in diff_files:
    diff = pd.read_csv(f, encoding='shift-jis', header=None, dtype=object)
    
    diff.columns = columns
    
    diff = diff.loc[ diff['prefectureName'] == '静岡県' ]
    data = pd.concat([data, diff])
    
print( len(data) )
```

```python
data.describe()
```

```python
print( data[data['corporateNumber'].duplicated()] )
```

```python
data.drop_duplicates(subset='corporateNumber', keep='last', inplace=True)
```
